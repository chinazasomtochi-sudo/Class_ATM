```python
class ATM:
    def __init__(self, initial_balance=0):
        self.balance = initial_balance
        
    def create_account(self):
        self.name = input("Enter your name: ")
        self.address = input("Enter your address: ")
        print(f"Account created for {self.name} living at {self.address}")
        
    def deposit(self):
        try:
            self.amount = float(input("Enter an amount to be deposited: "))
            if self.amount < 0:
                print("That is not a valid amount")
            else:
                self.balance += self.amount
                print(f"Deposited: #{self.amount:.2f}")
        except ValueError:
            print("Invalid input. Please enter a number.")
            
    def withdraw(self):
        try:
            self.amount = float(input("Enter an amount to be withdrawn: "))
            if self.amount > self.balance:
                print ("Insufficient funds")
            elif self.amount <= 0:
                print("Amount must be greater than 0")
            else:
                self.balance -= self.amount
                print(f"Withdrawn: #{self.amount:.2f}")
        except ValueError:
            print("Invalid input. Please enter a number.")
            
    def check_balance(self):
        print(f"Your balance is #{self.balance:.2f}")
atm = ATM()

while True:
    print("*******************************")
    print("Welcome to ATM")
    print("1. Create an account")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Check balance")
    print("5. Exit")
    print("*******************************")
    try:
        choice = int(input("Enter your choice between 1 and 5: "))
        print("*******************************")
        if choice == 1:
            atm.create_account()
        elif choice == 2:
            atm.deposit()
        elif choice == 3:
            atm.withdraw()
        elif choice == 4:
            atm.check_balance()
        elif choice == 5:
            print("Thank you.Bye!")
            break
        else:
            print("That is not a valid choice")
        print("*******************************")
    except ValueError: 
        print("Invalid input. Please enter a number between 1 and 5.")
        print("*******************************")
```
